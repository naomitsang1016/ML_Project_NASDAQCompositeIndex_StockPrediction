```python
df_list=[NASDAQ_Comp, Dow_Jones, S_P_500, AAPL, MSFT, AMZN, FB, GOOG, GOOGL, TSLA, NVDA, ASML, INTC, CMCSA, NFLX, ADBE, CSCO, PEP, AVGO, TXN, TMUS, _13_Week_Treasury_Bill, Treasury_Yield_5_Years, Treasury_Yield_10_Years, Treasury_Yield_30_Years, Gold, Oil, CNY_USD, GBP_USD, JPY_USD, EUR_USD, CAD_USD]
```

```python
for i in df_list:
    print(i.info())
```

```python
def dpna(a):
    for x in a:
        x.dropna(inplace=True)
        x.reset_index(drop=True,inplace=True)
    return a
    
df_list=dpna(df_list)    


def dr(a):
    for x in a:
        drp=[]
        yes=[]
        x['Open'].astype(float)
        for i in range(len(x['Open'])): 
            
            if i>0:
                drp.append((x['Open'][i]-x['Open'][i-1])/x['Open'][i-1]*100)
                if (x['Open'][i]-x['Open'][i-1])>0:
                    yes.append(1)
                else:
                    yes.append(0)
            else:
                drp.append(0)
                yes.append(np.nan)

        x['%_Daily_Return']=drp
        x['Yesterday_Price_Change']=yes
        
        
    return a

df_list=dr(df_list) 

def tr(a):
    for x in a:
        dc=[]
        dcp=[]
        tmr=[]
        for i in range(len(x['Open'])): 
            
            if i!=(len(x['Open'])-1):
                dcp.append((x['Open'][i+1]-x['Open'][i])/x['Open'][i])
                if (x['Open'][i+1]-x['Open'][i])>0:
                    dc.append(1)   
                else:
                    dc.append(0)
                if (x['Close'][i+1]-x['Close'][i])>0:
                    tmr.append(1)
                else:
                    tmr.append(0)
                
            else:
                dcp.append(np.nan)
                dc.append(np.nan)
                tmr.append(np.nan)
        
        x['%_Daily_Change']=dcp
        x['Daily_Change']=dc
        x['tmr_Change']=tmr
        
        
    return a

df_list=tr(df_list) 
```

```python
def colname(a,b):
    count=0
    for x in a:
        col_name=['Date']
        for i in range(1,len(x.columns)):
            col_name.append(b[count]+'_'+x.columns[i])
        x.columns=col_name
        count+=1
    return a
df_list=colname(df_list,list_df)     
```

```python

dfs = df_list
dfs = [df.set_index('Date') for df in dfs]
df=pd.DataFrame().join(dfs, how="outer")
```

```python
def MonthToSeason(x):   
        if x == 6 or x == 7 or x == 8 :
             return 'Summer'
        elif x == 10 or x == 11 or x == 9:
             return 'Autumn'
        elif x == 12 or x == 1 or x == 2:
             return 'Winter'
        elif x == 3 or x == 4 or x == 5:
             return 'Spring'
        else:
             season = np.nan 
        return season

df['Season'] = df['Date'].dt.month.apply(lambda x : MonthToSeason(x))
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(df[['Season']])
df_season_1hot = pd.DataFrame(encoder.transform(df[['Season']]).toarray(),columns=encoder.categories_)
df=pd.concat([df,df_season_1hot], axis=1)
```
